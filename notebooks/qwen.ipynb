```python
from openai import OpenAI
from IPython.display import Markdown, display
client = OpenAI(
    api_key="YOUR_API_KEY_HERE",
    base_url="http://192.168.68.70:8000/v1"
)
client.models.list()
```

```python
response = client.chat.completions.create(
    model="Qwen3-VL-32B-Thinking-Q8_0.gguf", # Specify the model name the server hosts
    messages=[
        {
            "role": "user",
            "content": [
                {"type": "text", "text": "What is in this image?"},
                {
                    "type": "image_url",
                    "image_url": {
                        "url": "https://upload.wikimedia.org/wikipedia/commons/f/fd/PaeoniaSuffruticosa7.jpg",
                    },
                },
            ],
        }
    ],
    max_tokens=300,
)

response_message = response.choices[0].message
display(Markdown(response_message.content))
```
