```python
from scipy.interpolate import CubicSpline
import numpy as np
from matplotlib import pyplot as plt

raw = "-168.427772082409,96.2735838096721,4.07972431182861;-159.821304073235,-4.2176526877434,2.43307089805603;-147.757427328512,-23.4428512912028,2.13156175613403;-133.34473680048,-27.8004384045588,2.14206027984619;-117.33829827654,-17.7245944623032,2.63418626785278;-110.4681619245,-5.92753414192727,3.03871393203735;-107.708601059204,15.9750359073449,3.39009189605713;-112.294365900106,72.976127283281,3.74081373214722"

pts = np.array([[float(v) for v in p.split(',')] for p in raw.split(';')])
x, y, z = pts[:,0], pts[:,1], pts[:,2]

# Parameterize by cumulative chord length
d = np.sqrt(np.diff(x)**2 + np.diff(y)**2)
t = np.concatenate([[0], np.cumsum(d)])
t /= t[-1]  # normalize to [0,1]

# Natural cubic spline (bc_type='natural' = zero second derivative at endpoints)
cs_x = CubicSpline(t, x, bc_type='natural')
cs_y = CubicSpline(t, y, bc_type='natural')

t_fine = np.linspace(0, 1, 500)
x_smooth = cs_x(t_fine)
y_smooth = cs_y(t_fine)
plt.figure(figsize=(8,6))
plt.plot(x, y, 'ro', label='Original Points')
plt.plot(x_smooth, y_smooth, 'b-', label='Cubic Spline')
plt.legend()
plt.title('Cubic Spline Interpolation of 2D Points')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis('equal')
plt.grid()
plt.show()
```
